```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('P05.csv')
```

```python
# Subsoal 1
# Menghitung distribusi jumlah kamar mandi di setiap kota

df.groupby('city')['bathroom'].count().plot(kind='bar', color='red')
plt.xlabel('Kota')
plt.ylabel('Jumlah Data Kamar Mandi')
plt.title('Distribusi Jumlah Kamar Mandi di Setiap Kota')
plt.show()
```

```python
df.groupby('animal')['rent_amount'].mean().plot(kind='bar', color=['lightgreen', 'lightcoral'])
plt.xlabel('Status Izin Hewan Peliharaan')
plt.ylabel('Rata-Rata Harga Sewa')
plt.title('Perbandingan Rata-Rata Harga Sewa Rumah dengan atau tanpa Hewan Peliharaan')
plt.show()
```

```python
df['rooms'].plot(kind="hist", bins=20, color='lightblue', rwidth = 0.9)
plt.xlabel('Jumlah Kamar Tidur')
plt.ylabel('Frekuensi')
plt.title('Histogram Distribusi Jumlah Kamar Tidur')
plt.show()
```

```python
# Perbandingan rata-rata biaya asuransi kebakaran dan luas tanah di setiap kota
avg_fire_insurance = df.groupby('city')['fire_insurance'].mean()
avg_area = df.groupby('city')['area'].mean()

# Visualisasi dalam bentuk bar chart
plt.figure(figsize=(10, 6))
avg_fire_insurance.plot(kind='bar', color='lightgreen', width=0.4, position=1, label='Rata-Rata Biaya Asuransi Kebakaran')
avg_area.plot(kind='bar', color='lightblue', width=0.4, position=0, label='Rata-Rata Luas Tanah')
plt.xlabel('Kota')
plt.ylabel('Rata-Rata')
plt.title('Perbandingan Biaya Asuransi Kebakaran dan Luas Tanah di Setiap Kota')
plt.legend
plt.show()
```
