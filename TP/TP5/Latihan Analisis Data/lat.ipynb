```python
import pandas as pd                 # Mengimport library pandas
import matplotlib.pyplot as plt     # Mengimport library matplotlib
df = pd.read_csv("data.csv")

df[["age"]].plot( kind = "hist", bins=[0, 20, 40, 60, 80, 100], rwidth=0.9)  # Membuat histogram data nilai matematika dengan step 10
plt.xlabel('Age of Person') 
```
