```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('P05-01.csv')
```

```python
# Subsoal 1

platinum_users = df[df['Card_Category'] == 'Platinum']
platinum_users['Customer_Age'].plot(kind='hist', bins=20, color='skyblue')
plt.title('Distribusi Usia Pengguna Kartu Platinum')
plt.xlabel('Usia Pengguna')
plt.ylabel('Frekuensi')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca data dari file CSV
df = pd.read_csv("P05-01.csv")

# Filter data untuk pengguna dengan limit kartu di atas 5000
limit_5000 = df[df['Credit_Limit'] > 5000]

# Menghitung jumlah pengguna berdasarkan kategori penghasilan
limit_5000['Income_Category'].value_counts().plot(kind='pie')
plt.title('Perbandingan Kategori Penghasilan Pengguna dengan Limit Kartu di atas 5000)')
plt.ylabel('')  # Menghilangkan label pada sumbu Y
plt.show()

```

```python
female_users = df[df['Gender'] == 'F']
female_users.plot(kind='scatter', x='Months_on_book', y='Avg_Utilization_Ratio')
plt.title('Scatter Plot: Lama Penggunaan vs Rata-rata Rasio Penggunaan\n(Pengguna Perempuan)')
plt.xlabel('Lama Penggunaan (Bulan)')
plt.ylabel('Rata-rata Rasio Penggunaan')
plt.show()
```

```python
status_counts = df['Attrition_Flag'].value_counts()
status_counts.plot(kind='barh', color='salmon')
plt.title('Jumlah Pengguna Berdasarkan Status Keaktifan')
plt.xlabel('Jumlah Pengguna')
plt.ylabel('Status Keaktifan')
plt.show()
```

```python
selected_marital_status = df[df['Marital_Status'].isin(['Married', 'Single', 'Divorced'])]
selected_marital_status.groupby(['Education_Level', 'Marital_Status']).size().unstack().plot(kind='bar', stacked=True)
plt.title('Status Perkawinan dan Jenjang Pendidikan Terakhir')
plt.xlabel('Jenjang Pendidikan')
plt.ylabel('Jumlah Pengguna')
plt.legend(title='Status Perkawinan')
plt.show()
```
