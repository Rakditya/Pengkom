```python
# SUBSOAL 1

import pandas as pd                 # Mengimport library pandas
import matplotlib.pyplot as plt     # Mengimport library matplotlib
df = pd.read_csv("P05.csv")

animal_allowed = df[df['animal'] == 'acept']
animal_allowed['rent_amount'].plot(kind= "hist", bins=20, color='skyblue', rwidth = 0.9)
plt.xlabel('Harga Sewa per Tahun')
plt.ylabel('Frekuensi')
plt.title('Distribusi Harga Sewa per Tahun dari Rumah yang Mengizinkan Hewan Peliharaan')
plt.show()
```

```python
# SUBSOAL 2

import pandas as pd                 # Mengimport library pandas
import matplotlib.pyplot as plt     # Mengimport library matplotlib
df = pd.read_csv("P05.csv")

bathroom_more_than_3 = df[df['bathroom'] > 3]
bathroom_more_than_3['city'].value_counts().plot(kind='pie')
plt.axis('equal')
plt.title('Perbandingan Kota Tempat Rumah Berada (Kamar Mandi > 3)')
plt.show()
```

```python
# SUBSOAL 3

import pandas as pd                 # Mengimport library pandas
import matplotlib.pyplot as plt     # Mengimport library matplotlib
df = pd.read_csv("P05.csv")

no_animal_allowed = df[df['animal'] == 'not acept']
no_animal_allowed.plot (kind='scatter', x='area', y='property_tax')
plt.xlabel('Luas Tanah')
plt.ylabel('Pajak Bangunan')
plt.title('Scatter Plot Luas Tanah vs Pajak Bangunan (Tanpa Izin Hewan Peliharaan)')
plt.show()
```

```python
# SUBSOAL 4

import pandas as pd                 # Mengimport library pandas
import matplotlib.pyplot as plt     # Mengimport library matplotlib
df = pd.read_csv("P05.csv")

bedroom_bathroom_filter = df[(df['rooms'] > 3) & (df['bathroom'] == 2)]
bedroom_bathroom_filter['city'].value_counts().plot(kind='barh', color='salmon')
plt.xlabel('Jumlah Rumah')
plt.ylabel('Kota')
plt.title('Jumlah Rumah di Setiap Kota (Kamar Tidur > 3 dan Kamar Mandi = 2)')
plt.show()
```

```python
# SUBSOAL 5

import pandas as pd                 # Mengimport library pandas
import matplotlib.pyplot as plt     # Mengimport library matplotlib
df = pd.read_csv("P05.csv")

furniture_city = df[df['city'].isin(['Campinas', 'Porto Alegre', 'Belo Horizonte'])]
furniture_city.groupby(['furniture', 'city']).size().unstack().plot(kind='bar', stacked=True)
plt.xlabel('Status Furniture')
plt.ylabel('Jumlah Rumah')
plt.title('Jumlah Rumah dengan Furniture di Beberapa Kota')
plt.legend(title='Kota')
plt.show()
```
